### **Applications**
Bessel functions are widely used in:
- Wave propagation (e.g., sound and electromagnetic waves in cylindrical systems).

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jn

x = np.linspace(0, 20, 10000)

plt.figure(figsize=(20, 6), dpi=350)
plt.plot(x, jn(1, x), label="j0(x)")  # 1st Order
plt.plot(x, jn(2, x), label="j1(x)")  # 2nd Order
plt.plot(x, jn(3, x), label="j3(x)")  # 3rd Order
plt.plot(x, jn(4, x), label="j4(x)")  # 4th Order

plt.legend()
plt.grid(True)
plt.show()
````

## Another Approach

````python
# Bessel's Function

import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jn

# Define the x range
x = np.linspace(0, 20, 10000)
# Plot Bessel functions of the first kind for different orders
m = np.arange(4.1)   #No of Orders
plt.figure(figsize=(20, 6), dpi=350)
for n in m:
    plt.plot(x, jn(n, x), label=f'J{n}(x)')
plt.title("Bessel Functions of the First Kind")
plt.xlabel("x")
plt.ylabel("J_n(x)")
plt.legend()
plt.grid(True)
plt.show()
````
