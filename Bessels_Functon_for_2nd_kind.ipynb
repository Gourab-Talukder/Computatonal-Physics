This code demonstrates plotting **Bessel functions of the second kind (denoted as \( Y_n(x) \))** for different orders. Below is an explanation for each section:

---

### **First Implementation**
#### **Code Highlights**
1. **Defining the \( x \)-Range**:
   ```python
   x = np.linspace(0, 20, 10000)
   ```
   - \( x \) ranges from \( 0 \) to \( 20 \) with 10,000 points for smoothness.
   - Starting at \( 0 \) for \( Y_n(x) \) is problematic since \( Y_n(x) \) has a singularity at \( x = 0 \).

2. **Plotting \( Y_n(x) \) for Specific Orders**:
   ```python
   plt.plot(x, yn(1, x), label="y0(x)")
   plt.plot(x, yn(2, x), label="y1(x)")
   plt.plot(x, yn(3, x), label="y3(x)")
   plt.plot(x, yn(4, x), label="y4(x)")
   ```
   - `yn` from `scipy.special` computes Bessel functions of the second kind for specific orders (1st to 4th).
   - Uses `plt.ylim(-1, 1)` to set the y-axis range for better visibility.

3. **Displaying the Plot**:
   - Adds a legend, gridlines, and adjusts figure size for better presentation.

---

### **Second Implementation**
#### **Code Highlights**
1. **Avoiding the Singularity**:
   ```python
   x = np.linspace(0.1, 20, 1000)
   ```
   - Starts \( x \) from \( 0.1 \) to avoid the singularity at \( x = 0 \), ensuring numerical stability.

2. **Dynamic Plotting for Multiple Orders**:
   ```python
   m = [0, 1, 2, 3, 4]
   for n in m:
       plt.plot(x, yn(n, x), label=f'Y{n}(x)')
   ```
   - Dynamically computes and plots \( Y_n(x) \) for orders \( n = 0 \) through \( n = 4 \).

3. **Customizing the Plot**:
   - Includes titles, labels, legends, gridlines, and limits for a clean presentation.
   - Adjusts figure size to \( (20, 8) \) for better visualization.

---

### **Graphical Differences**
1. **First Implementation**:
   - Directly specifies a few orders but starts at \( x = 0 \), leading to potential numerical instability due to singularity.
   - Focused on specific orders (1st through 4th).

2. **Second Implementation**:
   - Handles singularity by starting \( x \) at a small positive value (\( 0.1 \)).
   - More general and dynamic as it uses a loop for orders \( n = 0 \) to \( n = 4 \).

---

### **Output Description**
Both implementations produce oscillatory plots for \( Y_n(x) \), showcasing:
- Divergence as \( x \to 0^+ \) due to singularity.
- Oscillatory decay as \( x \to \infty \).

The second implementation provides a more robust and cleaner approach by avoiding numerical issues at \( x = 0 \).

---

### **Applications**
Bessel functions of the second kind are commonly used in:
- Wave propagation in cylindrical or spherical coordinates.
- Problems involving cylindrical waveguides.
- Acoustic or electromagnetic wave problems in non-homogeneous media.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import yn

x = np.linspace(0.1, 20, 10000)

plt.figure(figsize=(20, 8), dpi=350)
plt.plot(x, yn(1, x), label="y0(x)")  # 1st Order
plt.plot(x, yn(2, x), label="y1(x)")  # 2nd Order
plt.plot(x, yn(3, x), label="y3(x)")  # 3rd Order
plt.plot(x, yn(4, x), label="y4(x)")  # 4th Order
plt.ylim(-1, 1)
plt.legend()
plt.grid(True)
plt.show()
````

## Another Approach

````python
#Bessel;s Functon for 2nd kind
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import yn

# Define the x range, starting from a small positive value to avoid the singularity at x=0
x = np.linspace(0.1, 20, 1000)

# Plot Bessel functions of the second kind for different orders
plt.figure(figsize=(20, 8))

m = [0, 1, 2, 3, 4]
for n in m:
    plt.plot(x, yn(n, x), label=f'Y{n}(x)')
plt.ylim(-1, 1)
plt.title("Bessel Functions of the Second Kind")
plt.xlabel("x")
plt.ylabel("Yn(x)")
plt.legend()
plt.grid(True)
plt.show()
````
