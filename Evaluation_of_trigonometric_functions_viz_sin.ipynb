This code illustrates the series expansions for the sine and cosine functions using their Taylor series representations. Here's a breakdown of the components:

---

### **Key Components**

#### **1. Taylor Series for Sine and Cosine**

- **Sine Series**:
  \[
  \sin(x) = \sum_{n=0}^{m} \frac{(-1)^n x^{2n+1}}{(2n+1)!}
  \]
  - The function `sines` computes this series expansion for a given \( x \) and number of terms \( m \).

- **Cosine Series**:
  \[
  \cos(x) = \sum_{n=0}^{m} \frac{(-1)^n x^{2n}}{(2n)!}
  \]
  - The function `cosines` computes the cosine series expansion.

---

#### **2. Numerical Evaluation**

```python
x = np.linspace(-2*np.pi, 2*np.pi, 500)
m = 10
sine = sines(x, m)
cosine = cosines(x, m)
```
- **Range of \( x \)**:
  - Covers one full cycle of sine and cosine (\(-2\pi\) to \(2\pi\)).
- **Terms in the Series** (\( m \)):
  - \( m = 10 \) ensures good approximation accuracy.

---

#### **3. Exact Values**

```python
s = np.sin(np.pi/2)
c = np.cos(np.pi)
print("sin(π/2) =", s)
print("cosine(π) =", c)
```
- Computes the exact values of \( \sin(\pi/2) \) and \( \cos(\pi) \) for comparison:
  - \( \sin(\pi/2) = 1 \)
  - \( \cos(\pi) = -1 \)

---

#### **4. Plotting**

- **Sine Plot**:
  - Plots the sine series approximation.
  - Highlights the area under the curve using `fill_between`.

- **Cosine Plot**:
  - Similar treatment for the cosine series approximation.

```python
plt.subplot(2, 1, 1)
plt.plot(x, sine, color='k', label='Sine Expansion')
plt.fill_between(x, sine, color='m', alpha=0.15)

plt.subplot(2, 1, 2)
plt.plot(x, cosine, color='k', label='Cosine Expansion')
plt.fill_between(x, cosine, color='m', alpha=0.15)
```

---

### **Output**

- **Exact Values**:
  - \( \sin(\pi/2) = 1 \)
  - \( \cos(\pi) = -1 \)

- **Plots**:
  - The sine and cosine series expansions are plotted over the range \([-2\pi, 2\pi]\), showing good alignment with the respective functions.

---

### **Applications**

- **Numerical Approximation**:
  - Useful for approximating trigonometric functions in low-resource environments.
  
- **Fourier Analysis**:
  - Fundamental in signal processing and waveform analysis.

- **Visualization**:
  - Provides a clear view of the convergence behavior of Taylor series.

````python
import numpy as np
import matplotlib.pyplot as plt
import math

def sines(x, m):
    sum = 0.0
    for n in range(m):
        f = ((-1)**n) * (x**(2*n + 1)) / math.factorial(2*n + 1)
        sum = sum + f
    return sum

def cosines(x, m):
    sum = 0.0
    for n in range(m):
        f = ((-1)**n) * (x**(2*n)) / math.factorial(2*n)
        sum = sum + f
    return sum
x = np.linspace(-2*np.pi, 2*np.pi, 500)

m = 10
sine = sines(x, m)
cosine = cosines(x, m)
s = np.sin(np.pi/2)
c = np.cos(np.pi)
print("sin(π/2) =", s)
print("cosine(π) = ", c)

plt.figure(figsize=(10, 8))
plt.subplot(2, 1, 1)
plt.plot(x, sine, color = 'k', label = 'Sine Expansion')
plt.fill_between(x, sine, color = 'm', alpha = 0.15)
plt.xlabel('x')
plt.ylabel('Sine')
plt.legend()
plt.grid(True)
plt.title("Sine Series Expansion")

plt.subplot(2, 1, 2)
plt.plot(x, cosine, color = 'k', label = 'Cosine Expansion')
plt.fill_between(x, cosine, color = 'm', alpha = 0.15)
plt.xlabel('x')
plt.ylabel('Cosine')
plt.legend()
plt.grid(True)
plt.title("Cosine Series Expansion")
plt.tight_layout()
plt.show()
````
