```python
import numpy as np
import matplotlib.pyplot as plt

def plot_legendre(m):
    p0 = np.ones(len(x))  # P_0(x)
    p1 = x  # P_1(x)
    plt.plot(x, p0, label='P_0')
    plt.plot(x, p1, label='P_1')
    for n in range(2, m + 1):
        p3 = ((2 * n - 1) * x * p1 - (n - 1) * p0) / n
        plt.plot(x, p3, label=f'P_{n}')
        p0 = p1
        p1 = p3
    plt.xlabel('x')
    plt.ylabel('P_n(x)')
    plt.title('Legendre Polynomials Recurrence Relation')
    plt.legend(loc='lower right')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

x = np.linspace(-1, 1, 1000)
m = 4
plot_legendre(m)
```
