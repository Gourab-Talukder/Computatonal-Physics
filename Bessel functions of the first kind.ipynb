#### To plot and verify the recurrence relation of Bessel functions of the first kind
#### 1. Plotting Bessel Functions 
#### Generate values: x is generated over the interval [âˆ’20,20] with 1000 points.
#### Calculate Bessel functions: The Bessel functions J0(x), J1(x), J2(X) are calculated using scipy.special.jv().
#### Subplots:The code creates three subplots, each displaying one of the Bessel functions with labels, titles, legends, and grids for clarity.

```python
import numpy as np
import scipy.special as spl
import matplotlib.pyplot as plt

N = 1000
x = np.linspace(-20, 20, N)

j0, j1, j2 = spl.jv(0, x), spl.jv(1, x), spl.jv(2, x)

# Subplot 1: Bessel's function J_0(x)
plt.subplot(3, 1, 1)
plt.plot(x, j0, label='J0(x)')
plt.xlabel('x')
plt.ylabel("Bessel's function (J_0(x))")
plt.title("Bessel's Function J_0(x)")
plt.legend(loc ='upper right')
plt.grid(True)

# Subplot 2: Bessel's function J_1(x)
plt.subplot(3, 1, 2)
plt.plot(x, j1, label='J1(x)')
plt.xlabel('x')
plt.ylabel("Bessel's function (J_1(x))")
plt.title("Bessel's Function J_1(x)")
plt.legend(loc ='upper right')
plt.grid(True)

# Subplot 3: Bessel's function J_2(x)
plt.subplot(3, 1, 3)
plt.plot(x, j2, label='J2(x)')
plt.xlabel('x')
plt.ylabel("Bessel's function (J_2(x))")
plt.title("Bessel's Function J_2(x)")
plt.legend(loc ='upper right')
plt.grid(True)
plt.tight_layout()

n = 2
tol = 0.0001
left = (x * spl.jvp(n, x)) + (n * spl.jv(n, x))
right = x * spl.jv(n - 1, x)

if abs(left - right).max() < tol:
    print("Recurrence formula for Bessel function J_2(x) is verified.")
else:
    print("Recurrence formula for Bessel function J_2(x) is NOT verified.")

plt.plot(x, left, 'ro', markevery=8)
plt.plot(x, right, 'b-')
plt.title('Verification of recurrence formula for Bessel function')
plt.xlabel('x')
plt.tight_layout()
plt.show()
```
